## Q4（10点）：for loop

`nums` の合計を `total` に入れて print。

```python
# TODO: Q4
nums = [2, 4, 6, 8]

# ヒント:
# 1) total = 0 で初期化
# 2) for n in nums: でループ
# 3) total += n で足していく

# ここにコードを書いてください
total = 0
for n in nums:
    total += n
print(total)
```

## Q5（10点）：loop + if

`nums` から偶数だけを `evens` に集めて print。
（新しい list を作る）

```python
# TODO: Q5
nums = [1, 2, 3, 4, 5, 6]

# ヒント:
# 1) evens = [] で空のリストを作る
# 2) for n in nums: でループ
# 3) if n % 2 == 0: で偶数かチェック
# 4) evens.append(n) で追加

# ここにコードを書いてください
evens = []
for n in nums:
    if n % 2 == 0:
        evens.append(n)
print(evens)
```

## Q6（10点）：Functions

`mean_list(lst)` を作ってください。
- 引数：数値の list
- 戻り値：平均

最後に `mean_list([10, 20, 30])` を実行して結果を print。

```python
# TODO: Q6
# ヒント:
# 1) def mean_list(lst): で関数を定義
# 2) sum(lst) でリストの合計を計算
# 3) len(lst) でリストの長さを取得
# 4) return で結果を返す

# ここにコードを書いてください
def mean_list(lst):
    return sum(lst) / len(lst)
print(mean_list([10, 20, 30]))  # テストコード
```

```python
# === Auto-score setup ===
score = 0
max_score = 0

def award(name, ok, pts):
    global score, max_score
    max_score += pts
    if ok:
        score += pts
        print(f"[{name}] +{pts} pts")
    else:
        print(f"[{name}] +0 / {pts} pts")

def show_total():
    print("\n=== TOTAL ===")
    print("score:", score)
    print("max:", max_score)
```
